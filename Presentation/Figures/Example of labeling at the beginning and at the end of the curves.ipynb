```python
def ratioplotll(assets, start):
    from IPython.core.interactiveshell import InteractiveShell
    InteractiveShell.ast_node_interactivity = "all"    
    %matplotlib inline
    import matplotlib.pyplot as plt
    import matplotlib as mpl
    mpl.rcParams['axes.prop_cycle'] = mpl.cycler(color= ['r', 'coral', 'orange', 'lime', 'gold', 'green', 'darkgreen', 'cyan', 'magenta', 'blue', 'purple', 'black', 'brown',  'teal', 'turquoise', 'pink'])
    plt.tight_layout()
    from yahoofinancials import YahooFinancials
    import pandas as pd
    import numpy as np
    data = YahooFinancials(assets).get_historical_price_data(start_date=start, 
                                                      end_date='2021-03-19', 
                                                      time_interval='daily')
    prices_df = pd.DataFrame({
        a: {x['formatted_date']: x['adjclose'] for x in data[a]['prices']} for a in assets
    })
    fig, ax = plt.subplots()
    (prices_df / prices_df.iloc[0]).plot(ax=ax,figsize=(19,10), grid=True);
    for line, name in zip(ax.lines, list(prices_df.columns)):
        y = line.get_ydata()[-1]
        x = line.get_xdata()[-1]
        if not np.isfinite(y):
            y=next(reversed(line.get_ydata()[~line.get_ydata().mask]),float("nan"))
        if not np.isfinite(y) or not np.isfinite(x):
            continue        
        text = ax.annotate(name,
                           xy=(x, y),
                           xytext=(0, 0),
                           color=line.get_color(),
                           xycoords=(ax.get_xaxis_transform(),
                                     ax.get_yaxis_transform()),
                           textcoords="offset points")
        text_width = (text.get_window_extent(
            fig.canvas.get_renderer()).transformed(ax.transData.inverted()).width)
        if np.isfinite(text_width):
            ax.set_xlim(ax.get_xlim()[0], text.xy[0] + text_width * 1.05)    
    plt.tight_layout()
    plt.show();
    fig, ax = plt.subplots()
    (prices_df.iloc[-1] / prices_df).plot(ax=ax,figsize=(19,10), grid=True);
    for line, name in zip(ax.lines, list(prices_df.columns)):
        y = line.get_ydata()[0]
        x = line.get_xdata()[0]
        if not np.isfinite(y):
            y=next(line.get_ydata()[~line.get_ydata().mask],float("nan"))
        if not np.isfinite(y) or not np.isfinite(x):
            continue        
        text = ax.annotate(name,
                           xy=(x, y),
                           xytext=(0, 0),
                           color=line.get_color(),
                           xycoords=(ax.get_xaxis_transform(),
                                     ax.get_yaxis_transform()),
                           textcoords="offset points")
        text_width = (text.get_window_extent(
            fig.canvas.get_renderer()).transformed(ax.transData.inverted()).width)
        if np.isfinite(text_width):
            ax.set_xlim(ax.get_xlim()[0], text.xy[0] + text_width * 1.05)    
    plt.tight_layout()
    plt.show();    
```

```python
def ratioplotll(assets, start):
    from IPython.core.interactiveshell import InteractiveShell
    InteractiveShell.ast_node_interactivity = "all"    
    %matplotlib inline
    import matplotlib.pyplot as plt
    import matplotlib as mpl
    mpl.rcParams['axes.prop_cycle'] = mpl.cycler(color= ['r', 'coral', 'orange', 'lime', 'gold', 'green', 'darkgreen', 'cyan', 'magenta', 'blue', 'purple', 'black', 'brown',  'teal', 'turquoise', 'pink'])
    plt.tight_layout()
    from yahoofinancials import YahooFinancials
    import pandas as pd
    import numpy as np
    data = YahooFinancials(assets).get_historical_price_data(start_date=start, 
                                                      end_date='2021-03-19', 
                                                      time_interval='daily')
    prices_df = pd.DataFrame({
        a: {x['formatted_date']: x['adjclose'] for x in data[a]['prices']} for a in assets
    })
    fig, ax = plt.subplots()
    (prices_df / prices_df.iloc[0]).plot(ax=ax,figsize=(19,10), grid=True);
    for line, name in zip(ax.lines, list(prices_df.columns)):
        y = line.get_ydata()[-1]
        x = line.get_xdata()[-1]
        if not np.isfinite(y):
            y=next(reversed(line.get_ydata()[~line.get_ydata().mask]),float("nan"))
        if not np.isfinite(y) or not np.isfinite(x):
            continue        
        text = ax.annotate(name,
                           xy=(x, y),
                           xytext=(0, 0),
                           color=line.get_color(),
                           xycoords=(ax.get_xaxis_transform(),
                                     ax.get_yaxis_transform()),
                           textcoords="offset points")
        text_width = (text.get_window_extent(
            fig.canvas.get_renderer()).transformed(ax.transData.inverted()).width)
        if np.isfinite(text_width):
            ax.set_xlim(ax.get_xlim()[0], text.xy[0] + text_width * 1.05)    
    plt.tight_layout()
    plt.show();
    fig, ax = plt.subplots()
    (prices_df.iloc[-1] / prices_df).plot(ax=ax,figsize=(19,10), grid=True);
    for line, name in zip(ax.lines, list(prices_df.columns)):
        y = line.get_ydata()[0]
        x = line.get_xdata()[0]
        if not np.isfinite(y):
            y=next(iter(line.get_ydata()[~line.get_ydata().mask]),float("nan"))
        if not np.isfinite(y) or not np.isfinite(x):
            continue
        text_width = (ax.get_window_extent(
            fig.canvas.get_renderer()).transformed(ax.transData.inverted()).width) 
        text = ax.annotate(name,
                           xy=(x, y),
                           xytext=(-text_width/10, 0),
                           color=line.get_color(),
                           xycoords=(ax.get_xaxis_transform(),
                                     ax.get_yaxis_transform()),
                           textcoords="offset points")
    plt.tight_layout()
    plt.show();    
```
